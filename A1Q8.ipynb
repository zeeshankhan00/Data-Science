```python
from matplotlib import pyplot as plt
year = [2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012]
no_of_applicants = [921261, 929198, 1043739, 1186454,
1194938, 1304495, 1356805, 1282000, 479651]

plt.plot(year,no_of_applicants, color = 'green', marker="*",mec='red', linestyle = 'solid')
plt.title("No of applicants")
plt.xlabel('year', color='green')
plt.ylabel('Billion in $')
plt.show()
        
```
